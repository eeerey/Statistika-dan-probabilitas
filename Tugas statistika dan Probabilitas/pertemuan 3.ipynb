```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Parameter distribusi
mean = 5
std_dev = 2
sample_size = 1000

#Bangkitkan data tinggi badan
heights = np.random.normal(mean, std_dev, sample_size)

#Buat dataframe
df = pd.DataFrame(heights, columns=['Hari'])

#Tampilkan 5 baris pertama
print(df.head())

#Plot histogram
plt.figure(figsize=(8,6))
sns.histplot(df['Hari'], kde=True, color='blue')
plt.title('Distribusi waktu pengiriman paket')
plt.xlabel('Hari')
plt.ylabel('Frekuensi')
plt.show()

#Hitung probabilitas
prob = np.sum((df['Hari'] >= 3)) / sample_size
print(f"Probabilitas bahwa paket akan dikirim dalam waktu kurang dari 3 hari: {prob:.2f}")

#Hitung probabilitas
prob = np.sum((df['Hari'] >= 4) & (df['Hari'] <= 7)) / sample_size
print(f"Probabilitas bahwa paket akan dikirim dalam waktu antara 4 dan 7 hari: {prob:.2f}")

#Plot CDF
plt.figure(figsize=(8, 6))
sns.ecdfplot(df['Hari'], color='green')
plt.xlabel('Hari')
plt.ylabel('Probabilitas Kumulatif')
plt.show()

#Statistik deskriptif
print(df['Hari'].describe())
```
