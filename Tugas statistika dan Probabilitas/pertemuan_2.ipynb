```python
import numpy as np
import matplotlib.pyplot as plt

# Menghitung distribusi normal
mean = 50
std_dev = 5
data = np.random.normal(mean, std_dev, 100)

# Menghitung statistik
data_Mean = np.mean(data)
data_Median = np.median(data)
data_Std_dev = np.std(data)

# Menampilkan hasil statistik
print(f"mean : {data_Mean}")
print(f"median : {data_Median}")
print(f"std dev : {data_Std_dev}")

# Membuat plot data
plt.plot(data)

# Menambahkan titik untuk mean, median, dan standar deviasi
plt.scatter([0], [data_Mean], label="mean", color="yellow", zorder=5)
plt.scatter([0], [data_Median], label="median", color="red", zorder=5)

# Menambahkan garis untuk standar deviasi
plt.axhline(y=data_Mean + data_Std_dev, color="blue", linestyle="--", label="Mean + 1 SD")
plt.axhline(y=data_Mean - data_Std_dev, color="blue", linestyle="--", label="Mean - 1 SD")

# Menampilkan legenda dan plot
plt.legend()
plt.show()
```
