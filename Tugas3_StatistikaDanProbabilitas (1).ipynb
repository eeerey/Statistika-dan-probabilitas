```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Untuk membaca data dari file CSV
file_path = "C:\\Users\\Asus\\Downloads\\Data Sales.csv"  
df = pd.read_csv(file_path, encoding='windows-1252', delimiter=';')

# Melihat beberapa baris pertama untuk memastikan struktur data
print(df.head())

# Filter untuk kategori Furniture
df_furniture = df[df['Category'] == 'Furniture']

# Hitung frekuensi produk dalam kategori Furniture
frekuensi = df_furniture['Sub-Category'].value_counts()

# Menampilkan Frekuensi
print(frekuensi)

# Visualisasi Distribusi Frekuensi dengan Diagram Garis
plt.figure(figsize=(8, 4))

# Membuat diagram garis
sns.lineplot(x=frekuensi.index, y=frekuensi.values, marker='o', color='b')

# Menambahkan garis lurus vertikal dari titik di bawah setiap produk
for i, v in enumerate(frekuensi.values):
    plt.vlines(x=i, ymin=0, ymax=v, color='gray', linestyle='--', alpha=0.7)

# Menambahkan judul dan label
plt.title('Distribusi Frekuensi Produk Kategori Furniture', fontweight='bold')
plt.xlabel('Produk', fontweight='bold')
plt.ylabel('Frekuensi', fontweight='bold')
plt.xticks(rotation=90)  # Rotasi label produk jika diperlukan
plt.grid(True, axis='y', linestyle='-', alpha=0.7) # Menambah garis horizontal pada grafik
plt.show()
```

```python
import pandas as pd

# Membaca data dari file CSV
file_path = "C:\\Users\\Asus\\Downloads\\Data Sales.csv"  # Lokasi file CSV
data = pd.read_csv(file_path, encoding='windows-1252', delimiter=';')

# Melihat beberapa baris pertama untuk memastikan struktur data
print(data.head())

# Menyaring data untuk kategori 'Technology'
technology_data = data[data['Category'] == 'Technology']

# Menghitung total penjualan per produk berdasarkan 'Product Name'
product_sales = technology_data.groupby(['Sub-Category', 'Product Name'])['Quantity'].sum().reset_index()

# Mengurutkan data berdasarkan total penjualan terbanyak
top_products = product_sales.sort_values(by='Quantity', ascending=False)

# Menampilkan 5 produk dengan penjualan terbanyak
print("5 Produk dengan Penjualan Terbanyak dalam Kategori 'Technology':")
print(top_products.head())  # Menampilkan 5 produk teratas

# Jika ingin menampilkan produk terlaris dengan penjualan terbanyak
top_product_info = top_products.iloc[0]  # Produk dengan penjualan terbanyak
print("\nProduk dengan Penjualan Terbanyak:")
print(f"Sub-Category: {top_product_info['Sub-Category']}")
print(f"Nama Produk: {top_product_info['Product Name']}")
print(f"Total Penjualan: {top_product_info['Quantity']}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file CSV
file_path = "C:\\Users\\Asus\\Downloads\\Data Sales.csv"  # Lokasi file CSV
data = pd.read_csv(file_path, encoding='windows-1252', delimiter=';')

# Melihat beberapa baris pertama untuk memastikan struktur data
print(data.head())

# Agregasi data berdasarkan wilayah DKI Jakarta
jakarta_data = data[data['City'] == 'DKI Jakarta']  

# Agregasi data berdasarkan tahun dan kategori 
data_aggregated = jakarta_data.groupby(['Year', 'Category'])['Quantity'].sum().reset_index()

# Menampilkan data yang telah teragregasi
print(data_aggregated)

# Untuk mengatur warna visualisasi bar
SetWarna = ['#3DDAD7', '#2A93D5', '#135589']

# Membuat diagram batang
plt.figure(figsize=(10, 6))  # Menyesuaikan ukuran grafik
sns.barplot(x='Year', y='Quantity', hue='Category', data=data_aggregated, zorder=3, palette=SetWarna)

#Mengatur warna bar
# Menentukan jumlah kategori unik
unique_categories = data_aggregated['Category'].nunique()
if len(SetWarna) < unique_categories:
    SetWarna = sns.color_palette("pastel", unique_categories)

print("\n") 

# Menambahkan judul dan label
plt.title('Diagram Batang Pembelian Barang per Tahun di DKI Jakarta (2014 - 2017)', fontsize=14, fontweight='bold')
plt.xlabel('Tahun', fontsize=12, fontweight='bold')
plt.ylabel('Jumlah Pembelian Barang (unit)', fontsize=12, fontweight='bold')

# Menampilkan grid
plt.grid(True)

# Menampilkan grafik
plt.legend(title='Kategori Barang', bbox_to_anchor=(1.05, 1), loc='upper left')  # Menambahkan legenda
plt.tight_layout()  # Menyesuaikan tata letak agar tidak terpotong
plt.show()
```
